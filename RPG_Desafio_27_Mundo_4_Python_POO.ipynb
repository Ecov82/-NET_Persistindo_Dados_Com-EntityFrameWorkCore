## 🧩 Classe base: Personagem

Esta é a classe **abstrata conceitual** do jogo.
Ela representa a **ideia geral** de qualquer personagem.

```python
import random
from abc import ABC, abstractmethod

class Personagem(ABC):
    def __init__(self, nome, vida, ataque):
        self.nome = nome
        self.vida_max = vida
        self.vida = vida
        self.ataque = ataque

    @abstractmethod
    def atacar(self, alvo):
        pass

    def esta_vivo(self):
        return self.vida > 0
```

📌 Observe:
- A classe **não é o jogador**
- Ela é um **modelo genérico**
- Serve de base para herança

## 🦸 Classe Heroi (Herança + Polimorfismo)

O herói é um personagem, mas **ataca de forma diferente**.
Por isso, herdamos de Personagem e **sobrescrevemos** o método atacar().

```python
class Heroi(Personagem):
    def __init__(self, nome, vida, ataque):
        super().__init__(nome, vida, ataque)
        self.nivel = 1
        self.exp = 0

    def atacar(self, alvo):
        dano = random.randint(self.ataque, self.ataque + 5 + self.nivel)
        alvo.vida -= dano
        print(f'⚔️ {self.nome} (Nív {self.nivel}) atacou {alvo.nome} com força heroica! Dano: {dano}')

    def ganhar_exp(self, quantidade):
        self.exp += quantidade
        print(f'✨ {self.nome} ganhou {quantidade} de EXP!')
        if self.exp >= 10:
            self.subir_nivel()

    def subir_nivel(self):
        self.nivel += 1
        self.exp = 0
        self.ataque += 2
        self.vida_max += 10
        self.vida = self.vida_max
        print(f'🆙 LEVEL UP! {self.nome} subiu para o nível {self.nivel}!')
```

✅ Mesmo método (`atacar`)
✅ Comportamento diferente
👉 Isso é **polimorfismo**.

## 👹 Classe Monstro (Herança)

O monstro também é um personagem, mas seu ataque é mais simples.

```python
class Monstro(Personagem):
    def atacar(self, alvo):
        dano = random.randint(1, self.ataque)
        alvo.vida -= dano
        print(f'👹 {self.nome} atacou {alvo.nome}! Dano: {dano}')
```

## 🎮 Iniciando o jogo

Agora vamos criar os objetos (instâncias) do jogo.

```python
class Monstro(Personagem):
    def atacar(self, alvo):
        dano = random.randint(1, self.ataque)
        alvo.vida -= dano
        print(f'👹 {self.nome} atacou {alvo.nome}! Dano: {dano}')

heroi = Heroi('Artemis', 50, 10)

# Lista de possíveis monstros
def gerar_monstro():
    tipos = [
        {'nome': 'Goblin', 'vida': 20, 'ataque': 5},
        {'nome': 'Orc', 'vida': 40, 'ataque': 10},
        {'nome': 'Esqueleto', 'vida': 15, 'ataque': 8}
    ]
    escolha = random.choice(tipos)
    return Monstro(escolha['nome'], escolha['vida'], escolha['ataque'])

monstro = gerar_monstro()
print(f'🗡️ Um {monstro.nome} selvagem apareceu! Batalha iniciada!')
```

## 🔁 Loop de combate

Enquanto os dois estiverem vivos, a batalha continua.

```python
while heroi.esta_vivo() and monstro.esta_vivo():
    heroi.atacar(monstro)
    if monstro.esta_vivo():
        monstro.atacar(heroi)

    print(f'Vida do {heroi.nome}: {heroi.vida}')
    print(f'Vida do {monstro.nome}: {monstro.vida}')
    print('-' * 30)
```

## 🏁 Final da batalha

```python
if heroi.esta_vivo():
    print(f'🏆 {heroi.nome} venceu a batalha!')
    heroi.ganhar_exp(10)
else:
    print('💀 O herói foi derrotado...')
```

# ⚔️ O Grande Desafio Final
Nesta seção, consolidamos tudo o que aprendemos em um loop de jogo completo.

```python
import time

def iniciar_aventura():
    print('--- 🏰 BEM-VINDO AO REINO DE PYTHON ---')
    nome_heroi = input('Qual o nome do seu herói? ') or 'Artemis'
    meu_heroi = Heroi(nome_heroi, 60, 12)

    vitorias = 0

    while meu_heroi.esta_vivo() and vitorias < 3:
        inimigo = gerar_monstro()
        print(f'\n💥 RODADA {vitorias + 1}: Um {inimigo.nome} bloqueia seu caminho!')

        while meu_heroi.esta_vivo() and inimigo.esta_vivo():
            time.sleep(1)
            meu_heroi.atacar(inimigo)
            if inimigo.esta_vivo():
                inimigo.atacar(meu_heroi)

            print(f'❤️ {meu_heroi.nome}: {meu_heroi.vida}/{meu_heroi.vida_max} | ❤️ {inimigo.nome}: {max(0, inimigo.vida)}')

        if meu_heroi.esta_vivo():
            print(f'\n🏆 Você derrotou o {inimigo.nome}!')
            meu_heroi.ganhar_exp(10)
            vitorias += 1
        else:
            print('\n💀 Fim de jogo... Você lutou bravamente.')
            break

    if vitorias == 3:
        print('\n👑 PARABÉNS! Você limpou a masmorra e se tornou uma lenda!')

iniciar_aventura()
```

### 👑 O Desafio do Chefe Final
Adicionamos uma classe especializada para o Boss e um novo loop de jogo que inclui esta batalha épica.

```python
class Chefao(Monstro):
    def __init__(self, nome, vida, ataque):
        super().__init__(nome, vida, ataque)

    def atacar(self, alvo):
        dano = random.randint(self.ataque, self.ataque + 10)
        alvo.vida -= dano
        print(f'🔥 O CHEFÃO {self.nome} RUGE E ATACA! Dano massivo: {dano}')

def menu_jogador(heroi):
    print(f'\n--- Turno de {heroi.nome} (Vida: {heroi.vida}/{heroi.vida_max}) ---')
    print('[1] Atacar ⚔️')
    print('[2] Curar 🧪')
    escolha = input('Escolha sua ação: ')
    return escolha

def iniciar_campanha_final():
    print('--- 🏰 INICIANDO CAMPANHA: O DESPERTAR DO CHEFE ---')
    nome_heroi = input('Qual o nome do seu herói? ') or 'Artemis'
    meu_heroi = Heroi(nome_heroi, 80, 15)
    miraculo_usado = False

    vitorias = 0
    while meu_heroi.esta_vivo() and vitorias < 2:
        inimigo = gerar_monstro()
        print(f'\n⚔️ Inimigo {vitorias + 1}: {inimigo.nome} apareceu!')
        while meu_heroi.esta_vivo() and inimigo.esta_vivo():
            acao = menu_jogador(meu_heroi)
            if acao == '1':
                meu_heroi.atacar(inimigo)
            elif acao == '2':
                cura = random.randint(10, 20)
                meu_heroi.vida = min(meu_heroi.vida_max, meu_heroi.vida + cura)
                print(f'🧪 {meu_heroi.nome} recuperou {cura} de vida!')
            if inimigo.esta_vivo():
                inimigo.atacar(meu_heroi)
        if meu_heroi.esta_vivo():
            vitorias += 1
            meu_heroi.ganhar_exp(10)
        else: break

    if meu_heroi.esta_vivo():
        boss = Chefao('Dragão de Python', 150, 20)
        print(f'\n🌋 CUIDADO! {boss.nome} emergiu!')

        while meu_heroi.esta_vivo() and boss.esta_vivo():
            time.sleep(1)
            # --- SISTEMA DE MILAGRE CINEMATOGRÁFICO ---
            if meu_heroi.vida <= (meu_heroi.vida_max * 0.15) and not miraculo_usado:
                print(f'\n🎬 MÚSICA ÉPICA COMEÇA A TOCAR! 🎶')
                print(f'✨ {meu_heroi.nome} lembra de seus amigos e sente um poder ancestral!')
                print(f'⚡ MILAGRE: SEU ATAQUE FOI DOBRADO PARA O GOLPE FINAL! ⚡')
                meu_heroi.ataque *= 2
                miraculo_usado = True

            acao = menu_jogador(meu_heroi)
            if acao == '1':
                meu_heroi.atacar(boss)
            elif acao == '2':
                cura = random.randint(15, 25)
                meu_heroi.vida = min(meu_heroi.vida_max, meu_heroi.vida + cura)
                print(f'🧪 {meu_heroi.nome} se curou! +{cura} HP')

            if boss.esta_vivo():
                boss.atacar(meu_heroi)
            print(f'❤️ {meu_heroi.nome}: {meu_heroi.vida} | ❤️ BOSS: {max(0, boss.vida)}')

        if meu_heroi.esta_vivo():
            print(f'\n✨ LENDÁRIO! Com um golpe de misericórdia, {meu_heroi.nome} derrotou o Dragão!')
        else:
            print('\n💀 Você caiu... mas sua lenda será lembrada.')

iniciar_campanha_final()
```
